```python
Q3. What are the advantages and disadvantages of Elastic Net Regression?
```

#### The advantage of the elastic net is that it keeps the feature selection quality from the lasso penalty as well as the effectiveness of the ridge penalty. And it deals with highly correlated variables more effectively.

#### disadvantage of the classical elastic net is that the sequential cross-validation procedure used to determine the penalty parameters results in overshrinkage of the coefficients.

```python
Q4. What are some common use cases for Elastic Net Regression?
```

#### Empirically, elastic net is more stable than lasso when dealing with highly correlated features: suppose you have 2 features that are almost completely correlated. Depending on the noise, lasso will randomly select one or the other (provided that lambda is not too small and that they are both good predictors of the output) while elastic net will select both with approximately equals weights.

#### Use elastic net when you have several highly correlated variables. 

```python
Q5. How do you interpret the coefficients in Elastic Net Regression?
```

#### A positive coefficient indicates that as the value of the independent variable increases, the mean of the dependent variable also tends to increase. A negative coefficient suggests that as the independent variable increases, the dependent variable tends to decrease.
